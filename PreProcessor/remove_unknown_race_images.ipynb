```python
import os

from tqdm import tqdm
```

```python
RESOURCES_PATH="../resources/"
```

```python
def is_unknown_race(image):
    age, gender, race, time = image.split("_")
    return True if race == "4" else False
```

```python
def remove_unknown_race_images(directory):
    counter = 0
    for file in tqdm(os.listdir(directory)):
        if is_unknown_race(file):
            image_path = os.path.join(directory, file)
            print(f"is unknown race: {file}")
            counter = counter + 1
            os.remove(image_path)
    print(f"{counter} files removed.")
```

```python
remove_unknown_race_images(RESOURCES_PATH)
```
